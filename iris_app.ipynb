## Linear Regression:
### Model 1:

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

y = flowers[['SepalLengthCm']]
x = flowers[['SepalWidthCm']]

# Split the data into training and testing sets (30% test size)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
# Show the first five records of each dataset
x_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)

# # Show the first five records of actual and predicted values
y_test.head()
```

```python
pd.DataFrame(y_pred).head()
```

```python
# Calculate mean squared error
mse = mean_squared_error(y_test, y_pred)
print(f"\nMean Squared Error: {mse:.3f}")
```

### Model 2:

```python
y = flowers[['SepalLengthCm']]
x = flowers[['SepalWidthCm','PetalLengthCm','PetalWidthCm']]

# Split the data into training and testing sets (30% test size)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)

# Calculate mean squared error
mse = mean_squared_error(y_test, y_pred)
print(f"\nMean Squared Error: {mse:.3f}")
```

## Conclusion
I would believe that Model 2 is better as it uses more data to train and predict. As I see it, more data usually means better results. another thing is that the mean squared error is lower for Model 2(0.115) than that of Model 1(0.488) so that would mean that there is less room for error.
